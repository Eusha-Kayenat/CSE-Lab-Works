**`Alpha Beta Pruning - Coding Evaluation`**

```python
class Node:
    def __init__(self, value=None, children=None):
        self.value = value  # Value if it's a leaf node
        self.children = children if children is not None else [] # List of child nodes

# Level 3 (Leaf Nodes)
leaf_nodes_level3 = [
        Node(15), Node(15),
        Node(14), Node(-4),
        Node(0), Node(-14),
        Node(15), Node(-19)
    ]

# Level 2
node_level2_a = Node(children=[leaf_nodes_level3[0], leaf_nodes_level3[1]])
node_level2_b = Node(children=[leaf_nodes_level3[2], leaf_nodes_level3[3]])
node_level2_c = Node(children=[leaf_nodes_level3[4], leaf_nodes_level3[5]])
node_level2_d = Node(children=[leaf_nodes_level3[6], leaf_nodes_level3[7]])
nodes_level2 = [node_level2_a, node_level2_b, node_level2_c, node_level2_d]

# Level 1
node_level1_a = Node(children=[nodes_level2[0], nodes_level2[1]])
node_level1_b = Node(children=[nodes_level2[2], nodes_level2[3]])
nodes_level1 = [node_level1_a, node_level1_b]

# Level 0 (Root)
root = Node(children=[nodes_level1[0], nodes_level1[1]])

print("Tree structure based on the image created.")
```

```python
def print_tree(node, level=0):
    if node is not None:
        indent = "  " * level
        if node.value is not None:
            print(f"{indent}Value: {node.value}")
        else:
            print(f"{indent}Node:")
            for i, child in enumerate(node.children):
                print(f"{indent}  Child {i+1}:")
                print_tree(child, level + 2)

print("Printing the tree:")
print_tree(root)
```

**`Implement MinMax Algorithm with Alpha Beta Pruning`**

```python
import math
def alphabeta(node,depth,alpha,beta,maximizingPlayer):
  if depth==0:
    return node.value
  if maximizingPlayer:
    value=float('-inf')
    for i in node.children:
      value=max(value,alphabeta(i,depth-1,alpha,beta,False))
      if value>=beta:
       break
      alpha=max(value,alpha)
    return value
  else:
    value=float('inf')
    for i in node.children:
      value=min(value,alphabeta(i,depth-1,alpha,beta,True))
      if value<=alpha:
        break
      beta=min(beta,value)
    return value
alphabeta(root,3,float('-inf'),float('-inf'),True)
```
